## 單個顧客可能購買的單個商品precision/reacll/f1score計算

```python
#指定想預測的顧客，取出該顧客曾購買的產品紀錄(71519可替換成想預測的顧客ID)
customer_pass = train_data[train_data['customer']==71519 ]
customer_pass
```

```python
# 取出拿來驗證預測購買商品 precision/recall/f1score 的資料 (2021/8/-2021/12) => 71519可替換成想預測的顧客ID且需與上一欄相同
customer_example = test_data[test_data['customer']==71519]
customer_example
```

```python
# 取出想預測的顧客過去曾購買的所有產品
customer_pass_product = []
for i in range(len(customer_pass['product'])):
  customer_pass_product.append(customer_pass['product'].iloc[i])
```

```python
# 取出想預測的顧客驗證資料中購買的所有產品
example_product = []
for i in range(len(customer_example['product'])):
  example_product.append(customer_example['product'].iloc[i])
```

```python
# 去除重複購買的產品function
def uniq_product(products):
  b={}
  b=b.fromkeys(products)
  uniq_customer_pass_product=list(b.keys())
  return uniq_customer_pass_product
```

```python
# 針對單個商品算出與該商品最相近的十個商品作為未來可能購買的產品function
def sim_value(product):
  df_merge.loc[df_merge["product"] == product]
  product_embedding = df_merge.loc[df_merge["product"]==product, "vector"].iloc[0]
  df_merge["sim_value"] = df_merge["vector"].map(lambda x : 1 - distance.cosine(product_embedding, x))
  recom_ten = df_merge.sort_values(by="sim_value", ascending=False)[["product", "customer", "sim_value"]].head(10)
  rec_product = []
  for i in range(10):
    rec_product.append(recom_ten['product'].iloc[i])
  return rec_product
```

```python
# 想預測該名顧客曾購買的何項商品，並取出未重複產品
customer_pass_product_uniq = uniq_product(customer_pass_product)
example_product_uniq = uniq_product(example_product)
# 挑選任一個想預測該名顧客曾購買的商品(可將2984換成想測試的商品id)
product = 2984 
# 找出與上述挑的產品最相似的前十個
rec_product = sim_value(product)
# 針對"驗證資料中購買的所有商品"和"想預測的特定顧客曾購買的的單個特定產品(想挑選的商品)"做比對
repeat = []
for i in range(len(example_product_uniq)):
  if example_product_uniq[i] in rec_product:
    print("yes")
    repeat.append(example_product_uniq[i])
  else:
    print("no")
```

```python
if len(repeat)!=0:
  #precisin : 
  precision = len(repeat)/10
  #recall : 
  recall = len(repeat)/len(customer_pass_product_uniq)
  F1_score = 2 * precision * recall / (precision + recall)
else :
  precision = 0
  recall = 0
  F1_score = 0
# 得到precisin/recall/F1_score
print(precision)
print(recall)
print(F1_score)
```

## 多個顧客多個商品預測precision/reacll/f1score計算

```python
# 針對想預測的顧客驗證資料中購買的產品和過去曾購買的商品做比對
# 註 : 建議每次最多10000筆客戶資料，不然會執行很久(約1天)
# =>因此以下為針對all_customer資料的前10000筆客戶資料做購買預測的計算
all_precision = []
all_recall = []
all_f1score = []
for c in range(10000): # 前10000筆顧客，若想取得10000-20000的顧客資料可將10000改成10000,20000，以此類推
  id = all_customer['customer'].iloc[c]
  customer_pass = train_data[train_data['customer']==id]
  customer_example = test_data[test_data['customer']==id]
  customer_pass_product = []
  for i in range(len(customer_pass['product'])):
    customer_pass_product.append(customer_pass['product'].iloc[i])
  example_product = []
  for i in range(len(customer_example['product'])):
    example_product.append(customer_example['product'].iloc[i])

  customer_pass_product_uniq = uniq_product(customer_pass_product)
  example_product_uniq = uniq_product(example_product)
  all_repeat = []
  precision = []
  recall = []
  if len(customer_pass_product_uniq)!=0 and len(example_product_uniq)!=0:
    #某特定顧客過去購買
    for i in range(len(customer_pass_product_uniq)):
      rec_product = sim_value(customer_pass_product_uniq[i])
      repeat = []
      for j in range(len(example_product_uniq)):
        if example_product_uniq[j] in rec_product:
          print(example_product_uniq[j])
          repeat.append(example_product_uniq[j])
        else:
          print("no")
      if len(repeat)!=0:
        all_repeat.append(repeat)
      else:
        all_repeat.append(0)
      precision.append(len(repeat)/10)
      recall.append(len(repeat)/len(example_product_uniq))

    max_precision = max(precision, default=0)
    max_recall = max(recall, default=0)
    if (max_precision!=0 and max_recall!=0):
      f1_score = 2 * max_precision * max_recall / (max_precision + max_recall)
    else:
      f1_score = 0
  else:
    max_precision = 0
    max_recall = 0
    f1_score = 0
  
  all_precision.append(max_precision)
  all_recall.append(max_recall)
  all_f1score.append(f1_score)
```

```python
#新增要計算的欄位
all_customer['precision'] = 0
all_customer['recall'] = 0
all_customer['f1score'] = 0
```

```python
#若想取得10000-20000的顧客資料可將range(10000)改成range(10000,20000)/all_precision[i-10000(起始資料)]，以此類推
for i in range(10000):
  all_customer['precision'].iloc[i] = all_precision[i]
  all_customer['recall'].iloc[i] = all_recall[i]
  all_customer['f1score'].iloc[i] = all_f1score[i]
```

## t-SNE商品相關性視覺化

```python
word = model.getVectors().select("word")
word = np.array(word.select('word').collect())
#轉成一維
word = word.reshape(-1)

item_v = model.getVectors().select("vector")
item_v = np.array(item_v.select('vector').collect())
```

```python
word.shape
```

```python
item_v.shape
```

```python
item_v=item_v.reshape(item_v.shape[0],5)
```

```python
from sklearn import manifold
#t-SNE
X_tsne = manifold.TSNE(n_components=2, init='random', random_state=5, verbose=1).fit_transform(item_v)

#data Visualization 
x_min, x_max = X_tsne.min(0), X_tsne.max(0)
X_norm = (X_tsne - x_min) / (x_max - x_min) #normalize
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(40,40))
ax = fig.add_subplot()
for i in range(X_norm.shape[0]):
  ax.scatter(X_norm[i,0], X_norm[i,1])
  ax.annotate(word[i], (X_norm[i,0], X_norm[i,1]), fontproperties=myfont)

ax.set_title('axes title')

ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')

plt.show()
```
